**Дата:** 07/10/2024

**Тема:** Введение в деятельность, изучение математической индукции и основ по числовым последовательностям.

**Ход занятия:**

1. Прошли тестирование на знание Python.
2. Прошли тестирование на знание школьной математики.
3. Рассмотрели понятие математической индукции и способ её решения:
   - 3.1. Математическая индукция решается в 2 этапа:
     - 3.1.1. Рассмотрение базы при $$n = 1$$, после применяем правило: если за каждым верным утверждением следует верное утверждение, применяем математическую индукцию.
     - 3.1.2. Осуществление индукционного перехода (замена существующей переменной на другое обозначение, например: "k", прибавление к левой и правой части следующего члена, пример: "k + 1").
4. Рассмотрели один из примеров и доказали его равенство по математической индукции.
5. Рассмотрели основы математической последовательности:
   - 5.1. Дали определение
   - 5.2. Рассмотрели несколько способов задания (формулой общего члена, рекуррентный способ)
6. Заключительным этапом рассмотрели несколько видов числовых последовательностей:
   - 6.1. Постоянная
   - 6.2. Неограниченная
   - 6.3. Ограниченная
